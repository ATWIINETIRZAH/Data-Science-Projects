Now, we are going to handle the missing values depending on whether the data is categorical or numerical

```python
cat_features = df.select_dtypes(include=['object']).columns

num_features = df.select_dtypes(exclude=['object']).columns

# For categorical features, we shall fill NaN with the mode
df[cat_features] = df[cat_features].fillna(df[cat_features].mode())


# For numerical features we shall fill NaN with median
df[num_features] = df[num_features].fillna(df[num_features].median())
```

```python
print(df.isnull().sum().sum(), "missing values in the dataset")
```

```python
print(df.shape)
print(df.info())
```

```python
for col in num_features:
    df[col] = df[col].fillna(df[col].median())

for col in cat_features:
    df[col] = df[col].fillna("None")
```

```python
print(df.isnull().sum().sum(), "missing values in the dataset")
```
